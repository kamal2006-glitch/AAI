```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_derivative(y):
    return y * (1 - y)
def relu(x):
    return np.maximum(0, x)
def relu_derivative(x):
    return (x > 0).astype(float)
    def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)
def tanh(x):
    return np.tanh(x)
def tanh_derivative(x):
    return 1 - np.tanh(x)**2
def swish(x):
    return x / (1 + np.exp(-x))
def swish_derivative(x):
    sig = 1 / (1 + np.exp(-x))
    return sig + x * sig * (1 - sig)
def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)
def softmax_derivative(softmax_output):
    s = softmax_output.reshape(-1, 1)
    return np.diagflat(s) - np.dot(s, s.T)
X = np.array([[1, 1, 0, 1]])
Y = np.array([[1]])   
l = 0.71
t = 0.001
W1 = np.array([
    [0.3, 0.1, -0.2],
    [-0.2, 0.4, 0.3],
    [0.2, -0.3, 0.1],
    [0.1, 0.4, -0.1]
])
B1 = np.array([[0.2, 0.1, 0.05]])
W2 = np.array([
    [0.3, -0.2],
    [0.1, 0.4],
    [-0.3, 0.2]
])
B2 = np.array([[0.1, -0.2]])
V = np.array([
    [0.2],
    [-0.3]
])
C = np.array([[0.1]])
e = 0
p = False
while True:
    e += 1
    z1 = np.dot(X, W1) + B1
    h1 = relu(z1)
    z2 = np.dot(h1, W2) + B2
    h2 = relu(z2)
    u = np.dot(h2, V) + C
    o = sigmoid(u)
    d = Y - o
    do = d * sigmoid_derivative(o)
    dh2 = do.dot(V.T) * relu_derivative(z2)
    dh1 = dh2.dot(W2.T) * relu_derivative(z1)
    V += l * h2.T.dot(do)
    C += l * do
    W2 += l * h1.T.dot(dh2)
    B2 += l * dh2
    W1 += l * X.T.dot(dh1)
    B1 += l * dh1
    if not p:
        print("\n--- First Epoch ---")
        print("Output:", o.item())
        print("Error:", d.item())
        print("Updated W1:")
        for i in range(W1.shape[0]):
            for j in range(W1.shape[1]):
                print(f"  W1_{i+1}{j+1}: {W1[i,j]}")
        print("Updated b1:", B1.tolist())
        print("Updated W2:")
        for i in range(W2.shape[0]):
            for j in range(W2.shape[1]):
                print(f"  W2_{i+1}{j+1}: {W2[i,j]}")
        print("Updated b2:", B2.tolist())
        print("Updated V (Output weights):")
        for i in range(V.shape[0]):
            print(f"  V{i+1}: {V[i,0]}")
        print("Updated C (Output bias):", C.tolist())
        p = True
    if abs(d.item()) < t:
        print("\n--- Last Epoch ---")
        print("Epoch:", e)
        print("Output:", o.item())
        print("Error:", d.item())

        print("Final W1:")
        for i in range(W1.shape[0]):
            for j in range(W1.shape[1]):
                print(f"  W1_{i+1}{j+1}: {W1[i,j]}")

        print("Final b1:")
        for j in range(B1.shape[1]):
            print(f"  b1_{j+1}: {B1[0,j]}")

        print("Final W2:")
        for i in range(W2.shape[0]):
            for j in range(W2.shape[1]):
                print(f"  W2_{i+1}{j+1}: {W2[i,j]}")

        print("Final b2:")
        for j in range(B2.shape[1]):
            print(f"  b2_{j+1}: {B2[0,j]}")

        print("Final V (Output weights):")
        for i in range(V.shape[0]):
            print(f"  V{i+1}: {V[i,0]}")

        print("Final C (Output bias):")
        print(f"  C: {C[0,0]}")
        break
```
